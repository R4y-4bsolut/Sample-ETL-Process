## Transform Data

### 3.1. Enrollies Data Table Transform

```python
enrollies_data.info()
```

```python
enrollies_data['gender'] = enrollies_data['gender'].fillna(enrollies_data['gender'].mode()[0])
```

```python
enrollies_data.info()
```

```python
enrollies_data['full_name'] = enrollies_data['full_name'].astype('string')
```

```python
enrollies_data[['city', 'gender']] = enrollies_data[['city', 'gender']].astype('category')
```

```python
enrollies_data.info()
```

### 3.2. Enrollies' Education Table Transform

```python
enrollies_edu.info()
```

```python
enrollies_edu['enrolled_university'] = enrollies_edu['enrolled_university'].fillna(enrollies_edu['enrolled_university'].mode()[0])
```

```python
enrollies_edu['education_level'] = enrollies_edu['education_level'].fillna(enrollies_edu['education_level'].mode()[0])
```

```python
enrollies_edu['major_discipline'] = enrollies_edu['major_discipline'].fillna(enrollies_edu['major_discipline'].mode()[0])
```

```python
enrollies_edu[['enrolled_university', 'education_level', 'major_discipline']] = enrollies_edu[['enrolled_university', 'education_level', 'major_discipline']].astype('category')
```

```python
enrollies_edu.info()
```

### 3.3. Enrollies' Working Experience Table Transform

```python
enrollies_working_exp.info()
```

```python
enrollies_working_exp['experience'] = enrollies_working_exp['experience'].fillna(enrollies_working_exp['experience'].mode()[0])
```

```python
enrollies_working_exp['company_size'] = enrollies_working_exp['company_size'].fillna(enrollies_working_exp['company_size'].mode()[0])
```

```python
enrollies_working_exp['company_type'] = enrollies_working_exp['company_type'].fillna(enrollies_working_exp['company_type'].mode()[0])
```

```python
enrollies_working_exp['last_new_job'] = enrollies_working_exp['last_new_job'].fillna(enrollies_working_exp['last_new_job'].mode()[0])
```

```python
enrollies_working_exp[['relevent_experience', 'experience', 'company_size', 'company_type', 'last_new_job']] = enrollies_working_exp[['relevent_experience', 'experience', 'company_size', 'company_type', 'last_new_job']].astype('category')
```

```python
enrollies_working_exp.info()
```

## 4. Load Data

```python
from sqlalchemy import create_engine

db_path = 'data_warehouse.db'
engine = create_engine(f'sqlite:///{db_path}')
enrollies_data.to_sql('Dim_Enrollies', engine, if_exists='replace', index=False)
enrollies_edu.to_sql('Dim_Enrollies_Education', engine, if_exists='replace' , index=False)
enrollies_working_exp.to_sql('Dim_Work_Experience', engine, if_exists='replace', index=False)
training_hours.to_sql('Dim_Training_Hours', engine, if_exists='replace', index=False)
cities.to_sql('Dim_Cities', engine, if_exists='replace', index=False)
employment.to_sql('Fact_Employment', engine, if_exists='replace', index=False)
```
